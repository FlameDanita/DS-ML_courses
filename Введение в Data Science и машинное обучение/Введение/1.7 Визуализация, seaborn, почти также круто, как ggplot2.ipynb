```python
df_iris = pd.read_csv('iris.csv', index_col=0)
df_iris['petal length']
```

```python
for column in df_iris:
    if column != 'species':
        ax = sns.distplot(df_iris[column], label=column, kde=True)
        ax.legend()
```

```python
sns.violinplot(y=df_iris['petal length'])
```
